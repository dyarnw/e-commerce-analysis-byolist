```python
import pandas as pd 
import numpy as np 
import os 

# data
path_folder = r"D://NEW PORTFOLIO//olist_all datasets//all_dataset" # folder 
files = [f for f in os.listdir(path_folder) if f.endswith('.csv')] # get all files end in csv
# looping 
for file in files:
    full_path = os.path.join(path_folder, file)
    df = pd.read_csv(full_path, sep=None, engine = 'python', on_bad_lines='skip')

    print(f"loading dataset: {file}")
    print(f"shape(row, column):{df.shape}")

    display(df.head(5))
    print("-"*50)                                            
```

```python
data = {}
for file in files:
    full_path = os.path.join(path_folder, file)
    table_name = file.replace('.csv','')
    df = pd.read_csv(full_path, sep=None, engine='python', on_bad_lines='skip')
    df.columns = df.columns.str.replace('\ufeff','').str.strip().str.lower()
    data[table_name] = df
```

```python
for name, df in data.items():
    print(df.dtypes)
    print("-"*40)
```

```python
for table_name, df in data.items():
    
    obj_cols = df.select_dtypes(include=['object']).columns
    
    df[obj_cols] = df[obj_cols].apply(lambda x: x.astype(str).str.strip())

    num_str_cols = [c for c in df.columns if any(k in c.lower() for k in ['price', 'value'])]
    for col in num_str_cols:
        if df[col].dtype == 'object':
            df[col] = df[col].str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    
    for col in df.columns:
        time_keywords = ['_at', 'timestamp', 'date', 'delivery']
        if any(key in col.lower() for key in time_keywords):
            df[col] = pd.to_datetime(df[col], errors='coerce')
            
    print(f" all cleaning steps finished for {table_name}")
```

```python
for name, df in data.items():
    print(df.dtypes)
    print("-"*40)
```

```python
from sqlalchemy import types

sql_dtype_map = {}
for name, df in data.items():
    column_specs = {}
    for col in df.columns:
        if 'id' in col.lower():
            column_specs[col] = types.VARCHAR(50)
        elif any(key in col.lower() for key in ['city', 'state', 'category', 'name']):
            if 'state' in col.lower():
                column_specs[col] = types.VARCHAR(10)
            else:
                column_specs[col] = types.VARCHAR(100)
    sql_dtype_map[name] = column_specs
print("varchar for id and category done")
```

```python
print(sql_dtype_map['olist_orders_dataset'])
```

```python
for name, df in data.items():
    num_cols = df.select_dtypes(include=['number']).columns
    obj_cols = df.select_dtypes(include=['object']).columns
    data[name][num_cols] = df[num_cols].fillna(0)
    data[name][obj_cols] = df[obj_cols].fillna('Unknown')
```

```python
print("Checking Missing Values All Tables")
for name, df in data.items():
    total_null = df.isnull().sum().sum() # count all cell 
    # logic for status message
    if total_null == 0:
        status = "CLEAN"
    else:
        status = f"{total_null} NULLS"
    # formatting the output for a neat table look 
    print(f"{name:<40} | {status:<15}")

print("="*60)
```

```python
# translations
products = data['olist_products_dataset'].copy()
translation = data['product_category_name_translation'].copy()

translation.columns = translation.columns.str.replace('\ufeff','').str.strip()
products.columns = products.columns.str.strip()
# merge
try:
    products = pd.merge(products, translation, on='product_category_name', how = 'left')
    # use english
    products['product_category_name'] = products['product_category_name_english'].fillna(products['product_category_name'])
    
    products.drop('product_category_name_english', axis = 1, inplace=True)
    #
    data['olist_products_dataset'] = products
    display(data['olist_products_dataset'].head())
    
except KeyError as e:
    print(f"still keyerrorr: {e}")
```

```python
# checking translation 
null_count = data['olist_products_dataset']['product_category_name'].isnull().sum()
if null_count == 0:
    print("All translations were successful")
else:
    print(f"There are null {null_count} missing translations")
```

```python
# validation 
final_check = data['olist_products_dataset']['product_category_name'].isnull().sum()
print(f"blank: {final_check}")
if final_check == 0:
    print("data was cleaned")
```
